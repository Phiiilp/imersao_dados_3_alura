```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

url_dados = "https://github.com/alura-cursos/imersaodados3/blob/main/dados/dados_experimentos.zip?raw=true"

dados = pd.read_csv(url_dados, compression = "zip")
dados.tail()
```

```python
dados['tempo'].unique()
```

```python
dados['dose'].unique()
```

```python
dados['droga'].unique()
```

```python
dados['g-0'].unique()
```

```python
dados['tratamento'].value_counts().plot.pie(label='',labels=['Com droga','Com controle'],radius=1,title="Tratamento")
```

```python
dados['tempo'].value_counts().plot.bar(title='Horas gastar nos experimentos').set_xlabel('Horas')
dados['tempo'].value_counts().plot.bar(title='Horas gastar nos experimentos').set_ylabel("Experimentos")
```

```python
dados.columns = dados.columns.str.replace('-','')
dados
```

```python
dados.query('g1 > 0')
```

```python
dados['tratamento'].value_counts(normalize = True)
```

```python
dados['dose'].value_counts(normalize = True)
```

```python
mapa = {'droga': 'compostos'}
dados.rename(columns=mapa, inplace=True)
dados
```

```python
cod_compostos = dados['compostos'].value_counts().index[0:5]
cod_compostos
```

```python
dados.query('compostos in @cod_compostos')
```

```python

sns.set()
plt.figure(figsize=(11,9))
ax = sns.countplot(x='compostos',data=dados.query('compostos in @cod_compostos'), order=cod_compostos) 
ax.set_title('Top 5 compostos', fontsize= 35)
ax.set_xlabel('Compostos', fontsize=20)
ax.set_ylabel('Quantidade',fontsize=20)
plt.show()
```

```python

plt.figure(figsize=[11,9])
ax = sns.histplot(data=dados, x='g0', element="poly")
ax.set_ylabel('Quantidade',fontsize=20)
ax.set_xlabel('g0',fontsize=20)
plt.show()
```

```python

dados_g = dados.loc[:,'g0':'g771'].describe().T['mean']

plt.figure(figsize=[11,9])
ax = sns.histplot(data=dados_g, element='poly')
ax.set_title('Media dos Gs',fontsize=30)
ax.set_xlabel('Media',fontsize=20)
ax.set_ylabel('Quantidade',fontsize=20)
plt.show()
```

```python
sns.boxplot(x='g1',y='tratamento',data=dados)
```

```python

plt.figure(figsize=[11,9])
ax = sns.boxplot(y='g0',x='tempo',data=dados)
ax.set_title('DiferenÃ§a de tempo entre os valores de g0', fontsize=25)
ax.set_xlabel('Horas', fontsize=15)
ax.set_ylabel('g0', fontsize=15)
plt.show()
```
